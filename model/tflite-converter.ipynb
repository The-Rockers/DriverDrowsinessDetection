# Import TensorFlow Lite

```
from google.colab import auth
auth.authenticate_user()
```

```
import google.cloud.storage as gcs
import tensorflow as tf
```

```
bucket_name = "antisomnus-bucket"
blob_name = "models/en_model_v0.h5"
```

```
client = gcs.Client()
bucket = client.bucket(bucket_name)
model = bucket.blob(blob_name)
model.download_to_filename('en_model_v0.h5')
```

# Convert the Keras model to a TensorFlow Lite model and write the .tflite file

```
tflite_model = tf.keras.models.load_model('en_model_v0.h5')
converter = tf.lite.TFLiteConverter.from_keras_model(tflite_model)
tflite_model = converter.convert()

file = open( 'en_model_v0.tflite' , 'wb' )
file.write( tflite_model )
```

# Check the output data type and shape.

```
import tensorflow as tf
import numpy as np

interpreter = tf.lite.Interpreter(model_path="en_model_v0.tflite")
interpreter.allocate_tensors()

print(interpreter.get_input_details()[0]['shape'])
print(interpreter.get_input_details()[0]['dtype'])

print(interpreter.get_output_details()[0]['shape'])
print(interpreter.get_output_details()[0]['dtype'])
```

# Upload to GCS Bucket

```
blob_name = "models/en_model_v0.tflite"
```

```
tflite_blob = bucket.blob(blob_name)
tflite_blob.upload_from_filename("en_model_v0.tflite")
print("File {} uploaded to {}.".format(blob_name,bucket_name))
```
